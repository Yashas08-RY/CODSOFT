```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
data = pd.DataFrame(california.data,columns=california.feature_names)
data["SalesPrice"]=california.target * 100000 #convert median value
data.head()
```

```python
print(data.shape)
print()
data.info()
print()
data.describe()
```

```python
data.isnull().sum()
```

```python
sns.pairplot(data.sample(2000),height=2.5)
plt.tight_layout()
```

```python
sns.histplot(data["SalesPrice"],kde=True)
print("Skewness:%f"% data['SalesPrice'].skew())
print("Kurtosis:%f"% data['SalesPrice'].kurt())
```

```python
fig, ax=plt.subplots()
ax.scatter(x=data['MedInc'],y=data['SalesPrice'])
plt.ylabel('SalesPrice',fontsize=13)
plt.xlabel('MedInc(Median Income)',fontsize=13)
plt.show()

fig, ax=plt.subplots()
ax.scatter(x=data['HouseAge'],y=data['SalesPrice'])
plt.ylabel('SalesPrice',fontsize=13)
plt.xlabel('HouseAge',fontsize=13)
plt.show()
```

```python
from scipy import stats
from scipy.stats import norm,skew

sns.histplot(data['SalesPrice'],kde=True)
(mu, sigma) = norm.fit(data['SalesPrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

plt.legend({mu: 'mu = {:.2f}'.format(mu), sigma: 'sigma = {:.2f}'.format(sigma)},loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

fig = plt.figure()
res = stats.probplot(data['SalesPrice'], plot=plt)
plt.show()
```

```python
data['SalesPrice'] = np.log1p(data['SalesPrice'])

sns.histplot(data['SalesPrice'],kde=True,stat= "density")
(mu, sigma) = norm.fit(data['SalesPrice'])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

plt.legend({mu: 'mu = {:.2f}'.format(mu), sigma: 'sigma = {:.2f}'.format(sigma)},loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

fig = plt.figure()
res = stats.probplot(data['SalesPrice'], plot=plt)
plt.show()
```

```python
plt.figure(figsize=(10,10))
cor=data.corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.Grays)
plt.show()
```

```python
cor_target = abs(cor["SalesPrice"])
relevant_features = cor_target[cor_target>0.1]
names = [index for index,value in relevant_features.items() if index != 'SalesPrice']

print(names)
print(len(names))
```

```python
from sklearn.model_selection import train_test_split

x = data.drop("SalesPrice",axis=1)
y = data["SalesPrice"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
predictions = lr.predict(x_test)

print('Actual value of the house (log-scale):-',y_test.iloc[3])
print('Predicted value of the house (log-scale):-',predictions[5])
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test,predictions)
rmse = np.sqrt(mse)
print('Root Mean Squared error(RMSE):',rmse)
```
