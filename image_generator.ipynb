```python
import torch
from diffusers import StableDiffusionPipeline

if torch.cuda.is_available():
  pipe = StableDiffusionPipeline.from_pretrained(
      "runwayml/stable-diffusion-v1-5",
      torch_dtype=torch.float16
  )
  pipe = pipe.to("cuda")
  prompt = "A majestic lion, with a flowing mane illuminated by the golden hour sun, stands regally on a rocky outcrop overlooking a vast savanna. The sky is a gradient of warm oranges and soft purples, with a few wispy clouds. The lion's eyes are keen and focused, conveying a sense of calm power. Photorealistic, ultra-detailed, cinematic lighting."
  image = pipe(prompt).images[0]
  image.save("lion.png")
  print("GENERATED IMAGE:")
  display(image)
else:
  print("WARNING!: this code requires a GPU to run quickly. Please use Google Colab and set runtime -> change runtime type to T4 GPU.")
```
